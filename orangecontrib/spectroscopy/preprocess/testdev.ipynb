```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
angle = np.pi/6 
print(angle)
```

```python
x = np.array([i for i in range(100)]) - 50
```

```python
y = x**2
```

```python
r = [np.tan(angle)*i for i in range(len(y)) ]
```

```python
r[-1]/len(r)
```

```python
plt.plot(x,y)
plt.plot(x,r)
```
